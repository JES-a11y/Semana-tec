## Uso de DataFrames
Los DataFrames son el tipo de objetos principales de Pandas. Soun una representación de una tabla. Se pueden acceder a las filas y las columnas
## Proceso para analizar un archivo
1. Cargar un archivo en un DataFrame.
2. Hacer una revision de los tipos de datos del archivo.
3. Revisar si existen datos nulos y datos duplicados en el archivo.
4. Eliminar inconsistencias.
5. Llevar a cabo el análisis apoyados de gráfica.

```python
# Medidad de tendencia central
print(f"promedio: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].mean()}")
print(f"mediana: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].median()}")
print(f"moda: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].mode()}")
# Medidas de dispersion
print(f"Desviación estandar: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].std()}")
print(f"Varianza: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].var()}")
# Medidas de posicionamiento
print(f"Mínimo: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].min()}")
print(f"Q1: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].quantile(0.25)}")
print(f"Q2: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].quantile(0.50)}")
print(f"Q3: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].quantile(0.75)}")
print(f"Máximo: {df[" Mining, Manufacturing, Utilities (ISIC C-E) "].max()}")
```

```python
# Cajas y bigotes (box and whiskens) -> Gráficas de casjas (boxplot) 
df[" Per capita GNI "].plot.box()
```

```python
df[" Per capita GNI "].dtype
```

```python
columnas_númericas = []
columnas_no_númericas = []
for col in df.columns:
    if df[col].dtype == "object":
        columnas_no_númericas.append(col)
    else:
        columnas_númericas.append(col)
print(f'Columnas númericas: {columnas_númericas}')
print(f'Columnas no númericas: {columnas_no_númericas}')
```

```python
for col in columnas_númericas:
    df[col].plot.box()
    plt.show()
```

```python
for col in columnas_númericas:
    df[col].plot.kde()
    plt.show()
```

```python
for col in columnas_númericas:
    df[col].plot.kde(title=col)
    plt.show()
```
